```python
#Purchasing Analysis (Age)

age_df=pd.DataFrame(df)
bins = [0.0, 9.9, 14.9, 19.9, 24.9, 29.9,34.9, 39.9, 999.]
age_bins=["<10", "10-14", "15-19", "20-24","25-29","30-34", "35-39","40+"]


 #age_demo_grp = age_df.grouby (pd.cut(age_df["Age", bins, labels = group_names))
age_df ["Age Demographics"] =pd.cut(age_df["Age"], bins, labels=age_bins)

age_df.head()
```

```python
age = df [["SN", "Age"]]

age1 = age.drop_duplicates()

age1 ["Age Demographics"] =pd.cut(age_df["Age"], bins, labels=age_bins)

age1 = age1.groupby("Age Demographics")

#del (age1 ["Age"])


resultdf = age1.count()

del (resultdf ["Age"])

resultdf = resultdf.rename(columns = {"SN": "Total Count"})

resultdf ["Percentage of Players"] = resultdf["Total Count"]/total_table.iloc[0,0]*100

resultdf.head()


#total_table["Total Players"]
```

```python
age = df [["Age", "Price"]]


age ["AgeGroup"] =pd.cut(age_df["Age"], bins, labels=age_bins)

age2 = age.groupby("AgeGroup")

Count_age = age2["Price"].count()
Count_age1 = age2["Price"].sum()
Count_sum = Count_age1/Count_age

age3 = df [["Age", "SN"]].drop_duplicates()

age3 ["AgeGroup"] =pd.cut(age_df["Age"], bins, labels=age_bins)

age4 = age3.groupby("AgeGroup")

Count_age2= Count_age1/age4["Age"].count()


Age_dict= {"Purchase Count" : Count_age, "Average Purchase Price": Count_sum, "Total Purchase Value": Count_age1, "Avg Total Purchase per Person": Count_age2}

AgeAnalysis = pd.DataFrame (Age_dict)

AgeAnalysis.head()
    
```
